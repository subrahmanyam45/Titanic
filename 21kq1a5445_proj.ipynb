## 4.Univariate analysis and data visualization

```python
t["survived"].value_counts()
# it gives the count of survived and non survived persons i.e 0-dead & 1-survived
```

```python
t["survived"].value_counts(normalize=True)*100
# it gives the percentage of survived and non survived persons i.e 0-death percentage & 1-survived percentage
```

```python
# 61% of people died in the incident &
# 38% of people survived in the incident
```

```python
sns.countplot(t,x="survived")
# graphical representation of survived data through count plot
```

```python
t["pclass"].value_counts()
# it shows the count of pclass i.e there are three pclasses they are 1,2,3
```

```python
t["pclass"].value_counts(normalize=True)*100
# it gives the percentage of pclasses 
```

```python
# 55 % people travelled in 3rd class
```

```python
t["pclass"].value_counts(normalize=True).sort_index(ascending=True)*100
# the above line gives the output in order because we used sort_index method
```

```python
sns.countplot(t,x="pclass")
#the graphical representation of pclass with the help of count plot
```

```python
t["sibsp"].value_counts()
# it gives the values of how many siblings/spouse are there in titanic
```

```python
t["sibsp"].value_counts(normalize=True)*100
# it gives the percentage of how many siblings/spouse are there in titanic
```

```python
# 68% of people travelled alone
```

```python
sns.countplot(t,x="sibsp")
# it gives the graphical representaion of sibsp data
```

```python
t["age"].describe()
# age is not a categorical so we have to use describe methhod to get all its info
```

```python
# min age of titanic passenger is 0.42 i.e 4 months
# max age of titanic passenger is 80
# avg age of the passengers is 29
```

```python
sns.histplot(t,x="age")
# it shows the histplot of age data
```

```python
sns.kdeplot(t,x="age")
# the graphical reprsentation for age by usng kde plot
```

```python
sns.kdeplot(t,x="age",hue="survived")
```

```python
t["fare"].describe()
```

```python
# the count of the fares is 891
# the avg fare is 32
# the highest fare is 512
# the lowset fare is 0
```

```python
sns.kdeplot(t,x="fare")
```

```python
t["embark_town"].value_counts(normalize=True)*100
# it shows from which town we have more percentage of passengers for titanic
```

```python
# we have 72% of passengers from southampton  &
# we have 8% of passengers from Queenstown
```

```python
sns.countplot(t,x="embark_town",hue="sex")
# it shows graphical representation of embark_town with hue with the help of countplot
```

## 4.Bivariate analysis

```python
sns.swarmplot(t,x='survived',y='age')
# passengers age group between 20 to 40 are died more
# passengers age group below 10 years are survived more
```

```python
sns.stripplot(t,x='survived',y='age')
plt.title("Age vs Survived")
# it shows how many are survived in the titanic with age and survived
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(t,x='embark_town',y='age',hue='survived')
```
